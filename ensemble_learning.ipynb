```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('adult.txt', header=None)
df
```

```python
df = pd.get_dummies(df, columns = [1, 3, 5, 6, 7, 8, 9, 13, 14], drop_first=True)
df
```

```python
X = df.values[:,:-1]
y = df.values[:,-1]
```

```python
# Question 1:
# hard voting classifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
log_clf = LogisticRegression(solver='lbfgs')
rnd_clf = RandomForestClassifier(n_estimators=100)
svm_clf = SVC(gamma='scale')
voting_clf = VotingClassifier(
    estimators=[('lr', log_clf), ('rf', rnd_clf), ('svc', svm_clf)],
    voting='hard')
```

```python
# measure the algorithm performance using 10-fold cross validation
from sklearn.model_selection import cross_val_score
for clf in (log_clf, rnd_clf, svm_clf, voting_clf):
    hard_voting_score = cross_val_score(voting_clf, X, y, scoring = 'accuracy', cv=10)
    print(clf.__class__.__name__, hard_voting_score.mean())
```

```python
# soft voting classifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
log_clf = LogisticRegression(solver='lbfgs')
rnd_clf = RandomForestClassifier(n_estimators=100)
svm_clf = SVC(gamma='scale', probability=True)
voting_clf = VotingClassifier(
    estimators=[('lr', log_clf), ('rf', rnd_clf), ('svc', svm_clf)],
    voting='soft')
```

```python
# measure the algorithm performance using 10-fold cross validation
from sklearn.model_selection import cross_val_score
for clf in (log_clf, rnd_clf, svm_clf, voting_clf):
    soft_voting_score = cross_val_score(voting_clf, X, y, scoring = 'accuracy', cv=10)
    print(clf.__class__.__name__, soft_voting_score.mean())
```

```python
# there is not much difference between the accuracy score of hard voting and soft voting
```

```python
# Question 2:
# decision tree clasifier to use in bagging
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
bag_clf = BaggingClassifier(
    DecisionTreeClassifier(), n_estimators=100,
    max_samples=500, bootstrap=True, n_jobs=-1)
```

```python
from sklearn.model_selection import cross_val_score
accuracy1 = cross_val_score(bag_clf, X, y, scoring = 'accuracy', cv=10)
accuracy1.mean()
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
bag_clf = BaggingClassifier(
    DecisionTreeClassifier(), n_estimators=100,
    max_samples=500, bootstrap=True, max_features = 5, n_jobs=-1)
```

```python
from sklearn.model_selection import cross_val_score
accuracy2 = cross_val_score(bag_clf, X, y, scoring = 'accuracy', cv=10)
accuracy2.mean()
```

```python
# the one trained on all features performs better than the other one trained with max_features set to 5
```

```python
# question 3:
# using GradientBoostingClassifier
# n_estimators=200, learning_rate=0.25
from sklearn.ensemble import GradientBoostingRegressor
gbrt = GradientBoostingRegressor(max_depth=2, n_estimators=200, learning_rate=0.25)
gbrt.fit(X, y)
gbrt.score(X,y)
```

```python
# n_estimators=500, learning_rate=0.5
from sklearn.ensemble import GradientBoostingRegressor
gbrt = GradientBoostingRegressor(max_depth=2, n_estimators=500, learning_rate=0.5)
gbrt.fit(X, y)
gbrt.score(X,y)
```

```python
# n_estimators=1000, learning_rate=1.0
from sklearn.ensemble import GradientBoostingRegressor
gbrt = GradientBoostingRegressor(max_depth=2, n_estimators=1000, learning_rate=1.0)
gbrt.fit(X, y)
gbrt.score(X,y)
```

```python
# n_estimators=1000 and learning_rate=1.0 give the best performance boost
```
